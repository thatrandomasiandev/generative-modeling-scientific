# Diffusion Benchmark

Train DDPM on a 2D Gaussian mixture and evaluate with MMD.

```python
from gen_sci.data.gaussian_mixture_dgp import GaussianMixtureDGPConfig, generate_gaussian_mixture_data, sample_gaussian_mixture
from gen_sci.diffusion.ddpm import sample_ddpm, train_ddpm
from gen_sci.diffusion.metrics import maximum_mean_discrepancy, mode_coverage

cfg = GaussianMixtureDGPConfig(n_samples=3000, seed=42)
data = generate_gaussian_mixture_data(cfg)
result = train_ddpm(data.samples, timesteps=40, epochs=60, device="cpu")
generated = sample_ddpm(result, n_samples=1000, seed=0)
reference = sample_gaussian_mixture(cfg, n_samples=1000)

print("MMD:", maximum_mean_discrepancy(generated, reference))
print("Mode coverage:", mode_coverage(generated, data.ground_truth["means"]))
```

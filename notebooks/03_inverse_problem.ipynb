# Inverse Problem Solvers

Compare Tikhonov, LASSO, and diffusion posterior sampling on compressed sensing.

```python
from gen_sci.data.inverse_problem_dgp import InverseProblemDGPConfig, generate_inverse_problem_data, generate_signal_corpus
from gen_sci.inverse.metrics import relative_l2_error
from gen_sci.inverse.solvers import diffusion_posterior_sample, lasso_solve, tikhonov_solve

cfg = InverseProblemDGPConfig(seed=42)
data = generate_inverse_problem_data(cfg)
corpus = generate_signal_corpus(cfg)

for name, result in [
    ("tikhonov", tikhonov_solve(data.A, data.y, lam=0.05)),
    ("lasso", lasso_solve(data.A, data.y, alpha=0.005)),
    ("diffusion", diffusion_posterior_sample(data.A, data.y, corpus.samples, n_samples=32, epochs=40)),
]:
    err = relative_l2_error(result.x_hat, data.x_true)
    print(f"{name}: rel_l2={err:.4f}")
```

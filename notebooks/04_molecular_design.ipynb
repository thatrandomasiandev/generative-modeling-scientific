# Molecular Design Search

Compare random, hill-climb, and guided diffusion search on a latent property landscape.

```python
from gen_sci.data.molecular_dgp import MolecularDGPConfig, generate_molecular_data
from gen_sci.molecular.metrics import regret
from gen_sci.molecular.search import guided_diffusion_search, hill_climb_search, random_search

cfg = MolecularDGPConfig(seed=42)
data = generate_molecular_data(cfg)
gt = data.ground_truth
budget = 200

for fn in [random_search, hill_climb_search]:
    res = fn(cfg.latent_dim, budget, gt["optimum"], gt["active_dims"], cfg.interaction_strength, seed=42)
    print(fn.__name__, "best=", res.best_property, "regret=", regret(res.best_property, gt["oracle_value"]))

diff = guided_diffusion_search(data.latents, budget, gt["optimum"], gt["active_dims"], cfg.interaction_strength, seed=42, epochs=40)
print("guided_diffusion", "best=", diff.best_property, "regret=", regret(diff.best_property, gt["oracle_value"]))
```
